## Model optimization

### Subtask:
Optimize the LSTM model's hyperparameters to improve its performance.


**Reasoning**:
Use Keras Tuner to search for optimal hyperparameters for the LSTM model, using F1-score as the evaluation metric and early stopping to prevent overfitting.



```python
import kerastuner as kt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import f1_score

def build_model(hp):
    model = Sequential()
    model.add(LSTM(units=hp.Int('units', min_value=25, max_value=100, step=25),
                   activation=hp.Choice('activation', values=['relu', 'tanh']),
                   input_shape=(X_train_reshaped.shape[1], X_train_reshaped.shape[2])))
    for i in range(hp.Int('num_layers', 1, 2)):
        model.add(LSTM(units=hp.Int(f'units_{i}', min_value=25, max_value=100, step=25),
                       activation=hp.Choice('activation', values=['relu', 'tanh'])))
        model.add(Dropout(hp.Float('dropout', min_value=0.1, max_value=0.3, step=0.1)))
    model.add(Dense(1, activation='sigmoid'))

    optimizer = hp.Choice('optimizer', values=['adam', 'rmsprop'])
    lr = hp.Float('learning_rate', min_value=1e-4, max_value=1e-2, sampling='log')
    if optimizer == 'adam':
        optimizer = 'adam'
    else:
        optimizer = 'rmsprop'
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    return model

tuner = kt.Hyperband(
    build_model,
    objective=kt.Objective('val_accuracy', direction='max'), # Using validation accuracy
    max_epochs=50,
    factor=3,
    directory='my_dir',
    project_name='intro_to_kt'
)

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

tuner.search(X_train_reshaped, y_train_reshaped, epochs=50, validation_data=(X_val_reshaped, y_val_reshaped), callbacks=[early_stopping])

best_hps=tuner.get_best_hyperparameters(num_trials=1)[0]
print(f"""
The hyperparameter search is complete. The optimal number of units in the first densely-connected
layer is {best_hps.get('units')}, in the second layer is {best_hps.get('units_1')}, the optimal learning rate for the optimizer
is {best_hps.get('learning_rate')}, and the optimal activation function is {best_hps.get('activation')}.
""")
best_model = tuner.get_best_models(num_models=1)[0]

# Evaluate the best model on the validation set.
y_pred_val = (best_model.predict(X_val_reshaped) > 0.5).astype("int32")
f1 = f1_score(y_val_reshaped, y_pred_val, average='weighted')
print(f"Validation F1-score: {f1}")

# Save the model
best_model.save('best_lstm_model.h5')
```

```python
import numpy as np
from sklearn.metrics import f1_score
import kerastuner as kt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense
from tensorflow.keras.callbacks import EarlyStopping


def create_sequences(X, y, seq_length):
    xs = []
    ys = []
    for i in range(len(X) - seq_length):
        x = X[i:(i + seq_length)]
        y_val = y[i + seq_length]  # Target is at the next timestep
        xs.append(x)
        ys.append(y_val)
    return np.array(xs), np.array(ys)

timesteps = 12
X_train_reshaped, y_train_reshaped = create_sequences(X_train.values, y_train.values, timesteps)
X_val_reshaped, y_val_reshaped = create_sequences(X_val.values, y_val.values, timesteps)

print("X_train_reshaped shape:", X_train_reshaped.shape)
print("y_train_reshaped shape:", y_train_reshaped.shape)
print("X_val_reshaped shape:", X_val_reshaped.shape)
print("y_val_reshaped shape:", y_val_reshaped.shape)

def build_model(hp):
    model = Sequential()
    model.add(LSTM(units=hp.Int('units', min_value=25, max_value=100, step=25),
                   activation=hp.Choice('activation', values=['relu', 'tanh']),
                   return_sequences=True,  # Important for stacked LSTMs
                   input_shape=(X_train_reshaped.shape[1], X_train_reshaped.shape[2])))
    for i in range(hp.Int('num_layers', 1, 2)):
        model.add(LSTM(units=hp.Int(f'units_{i}', min_value=25, max_value=100, step=25),
                       activation=hp.Choice('activation', values=['relu', 'tanh']),
                       return_sequences=True if i < 1 else False))  # Only return sequences for intermediate layers
        model.add(Dropout(hp.Float('dropout', min_value=0.1, max_value=0.3, step=0.1)))
    model.add(Dense(1, activation='sigmoid'))

    optimizer = hp.Choice('optimizer', values=['adam', 'rmsprop'])
    lr = hp.Float('learning_rate', min_value=1e-4, max_value=1e-2, sampling='log')
    if optimizer == 'adam':
      optimizer = 'adam'
    else:
      optimizer = 'rmsprop'
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    return model

tuner = kt.Hyperband(
    build_model,
    objective=kt.Objective('val_accuracy', direction='max'),
    max_epochs=50,
    factor=3,
    directory='my_dir',
    project_name='intro_to_kt'
)

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

tuner.search(X_train_reshaped, y_train_reshaped, epochs=50, validation_data=(X_val_reshaped, y_val_reshaped), callbacks=[early_stopping])

best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
print(f"""
The hyperparameter search is complete. The optimal number of units in the first densely-connected
layer is {best_hps.get('units')}, in the second layer is {best_hps.get('units_1')}, the optimal learning rate for the optimizer
is {best_hps.get('learning_rate')}, and the optimal activation function is {best_hps.get('activation')}.
""")

best_model = tuner.get_best_models(num_models=1)[0]
y_pred_val = (best_model.predict(X_val_reshaped) > 0.5).astype("int32")
f1 = f1_score(y_val_reshaped, y_pred_val, average='weighted')
print(f"Validation F1-score: {f1}")

best_model.save('best_lstm_model.h5')
```
